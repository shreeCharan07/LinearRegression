```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/SOCR-HeightWeight.csv')
df
```

```python
x = df[['Height(Inches)']].values
x
```

```python
y = df['Weight(Pounds)'].values
y
```

```python
plt.scatter(x[:50],y[:50])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
x.ndim, y.ndim
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0, test_size = 0.01)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
df1 = pd.DataFrame({'Actual' :y_test, 'Predicted' : y_pred})
```

```python
df1.plot(figsize=(50,10)) #pandas plotting
plt.show()
```

```python
sns.regplot(y_test, y_pred)
```
